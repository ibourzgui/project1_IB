```python
%matplotlib inline

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
file_path1 = "asthma_mort_states_2016.csv"
asthma_data = pd.read_csv(file_path1)

file_path2 = "AQI_data.csv"
chem_data = pd.read_csv(file_path2)
```

```python
asthma_data.head()
```

```python
# convert all year cols into numeric
asthma_data.dtypes
asthma_data = asthma_data.convert_objects(convert_numeric=True)
asthma_data.dtypes
```

```python
asthma_data.head()
```

```python
#groupby state
asthma_data.groupby("LocationDesc").count().head()
```

```python
# filter years 2010-2016 asthma mortality data
asthma_data = asthma_data[['LocationDesc', '2010','2011', '2012', '2013','2014','2015', '2016']]
asthma_data.head()
```

```python
#asthma_data.groupby("LocationDesc")[[2010,2011,2012, 2013,2014,2015,2016]].mean()
asthma_data.columns
```

```python
#melt year columns
asthma_melted_df=pd.melt(asthma_data, id_vars=['LocationDesc'], value_vars=['2010','2011','2012','2013','2014','2015','2016'])
asthma_melted_df.head()
```

```python
#rename columns
asthma_melted_df = asthma_melted_df.rename(index=str, columns={"LocationDesc": "State", "variable": "Year", "value":"Asthma Mortality"})
asthma_melted_df.head()
asthma_melted_df.to_csv('asthma_mort_states_melted.csv', index=True, header=True)
asthma_melted_df.head()
```

```python
# calculate mean asthma mortality of 2010-2016 by state
Avg_asthma_byState = pd.DataFrame(asthma_melted_df.groupby("State")["Asthma Mortality"].mean())
Avg_asthma_byState.head()
```

```python
#reset index
Avg_asthma_byState.reset_index(inplace=True)

Avg_asthma_byState.head()
```

```python
#remove United States row
Avg_asthma_noUS = Avg_asthma_byState.loc[Avg_asthma_byState['State'] != 'United States']
Avg_asthma_noUS.head()
```

```python
# plot avg asthma mortality for the period of 2010-2016 by state
Avg_asthma_noUS.plot(x='State', y='Asthma Mortality', kind="bar", figsize=(20,3))
plt.title("Avg Asthma Mortality per State(2010-2016)")
plt.xlabel("State")
plt.ylabel("Asthma Mortality Num")
plt.grid()

```

```python
#filter pollt AQI dataset by selected cols
chem_data.head()

chemfilt = chem_data[['State', 'Year','NO2 AQI', 'O3 AQI', 'SO2 AQI', 'CO AQI']]
chemfilt.head()
```

```python
#filter pollt AQI dataset by years 2010-2016

#chemfilt.dtypes
chemfilt['Year']=chemfilt['Year'].astype('str')
chemfilt.dtypes
chemfilt=chemfilt[chemfilt['Year'].str.contains('2010|2011|2012|2013|2014|2015|2016')]
chemfilt.head()
```

```python
# merge asthma mortality data with pollt AQIs
merge_final_df = pd.merge(chemfilt, Avg_asthma_byState, on =['State'])
merge_final_df.to_csv('AQI_asthma_mort_states_melted.csv', index=True, header=True)

merge_final_df.head()
#merge_final_df.boxplot(column='Asthma Mortality', by='Year')
```

```python
# grouped by year the merged df and calc mean (AQIs + Asthma motality 2010-2016)
avg_AQI_asthm_byYear_df = pd.DataFrame(merge_final_df.groupby('Year')['Asthma Mortality','NO2 AQI', 'O3 AQI', 'SO2 AQI', 'CO AQI'].mean())
avg_AQI_asthm_byYear_df.reset_index(inplace=True)

avg_AQI_asthm_byYear_df
```

```python
# average asthma mortality per year nationally in the US
avg_asth_byYear_df=pd.DataFrame(merge_final_df.groupby(['State','Year'])['Asthma Mortality'].mean())
avg_asth_byYear_df.reset_index(inplace=True)

avg_asth_byYear_df.head()
```

```python
# total asthma mortality per year nationally in the US

tot_asth_byYear_df=pd.DataFrame(avg_asth_byYear_df.groupby(['Year'])['Asthma Mortality'].sum())
tot_asth_byYear_df.reset_index(inplace=True)

tot_asth_byYear_df
```

```python
pollutants=list(avg_AQI_asthm_byYear_df.columns[2:])

pollutants
```

```python
#graphs
#total Asthma deaths per year
tot_asth_byYear_df.plot(x='Year', y='Asthma Mortality', kind="bar")
plt.title("Total National Asthma Mortality Over time")
plt.xlabel("Years")
plt.ylabel("Sum of Mortality")
plt.grid()




for pol in pollutants:    
    avg_AQI_asthm_byYear_df.plot(x='Year', y=pol, kind="bar")
    plt.title(pol)
    plt.xlabel("Year")
    plt.ylabel("AQI")
    plt.grid()
#, figsize=(20,3)
```

```python
# average Asthma moratlity in the US per year  
ax=sns.swarmplot(x='Year', y='Asthma Mortality', data=avg_asth_byYear_df, color="grey")

ax=sns.boxplot(x='Year', y='Asthma Mortality', data=avg_asth_byYear_df)

ax.set(ylim=(0, 500))
ax.set_title("National Average Asthma Mortality")
```

```python
# average Asthma moratlity in the US per year  
ax = sns.boxplot(x='Year', y='Asthma Mortality', data=avg_asth_byYear_df)
ax = sns.stripplot(x='Year', y='Asthma Mortality', data=avg_asth_byYear_df, color="orange", jitter=0.2, size=2.5)
plt.title("National Average Asthma Mortality", loc="center")
```
