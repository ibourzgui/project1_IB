```python
#melt columns
asthma_melted_df=pd.melt(asthma_data, id_vars=['LocationDesc'], value_vars=['2010','2011','2012','2013','2014','2015','2016'])
asthma_melted_df.head()
```

```python
#rename columns
asthma_melted_df = asthma_melted_df.rename(index=str, columns={"LocationDesc": "State", "variable": "Year", "value":"Asthma Mortality"})
asthma_melted_df.head()
asthma_melted_df.to_csv('asthma_mort_states_melted.csv', index=True, header=True)
asthma_melted_df.head()
```

```python
Avg_asthma_byState = pd.DataFrame(asthma_melted_df.groupby("State")["Asthma Mortality"].mean())
Avg_asthma_byState.head()
```

```python
#Avg_asthma_byState.reset_index(inplace=True)
Avg_asthma_byState.reset_index(inplace=True)

Avg_asthma_byState
```

```python
#remove United States row
Avg_asthma_noUS = Avg_asthma_byState.loc[Avg_asthma_byState['State'] != 'United States']
Avg_asthma_noUS.head()
# plt.plot(Avg_asthma_byState['State'], Avg_asthma_byState['Asthma Mortality'], marker='o')
# plt.title("Asthma vs State")
# plt.xlabel("State")
# plt.ylabel("Asthma")
# plt.grid()
```

```python
Avg_asthma_noUS.plot(x='State', y='Asthma Mortality', kind="bar")
plt.title("Asthma per State")
plt.xlabel("State")
plt.ylabel("Asthma Mortality Num")
plt.grid()

```

```python
#filter dataset by year:2010 to 2016
chem_data.head()
cfilt=chem_data
cfilt=cfilt[cfilt['Year'].str.contains('2010|2011|2012|2013|2014|2015|2016')]

cfilt.head()
```

```python
cfilt['Year']=cfilt['Year'].astype('str')
cfilt.drop(['State Average'], axis=1)
statesummary=pd.DataFrame(cfilt.melt(id_vars=['Year','State']).groupby(['State','variable'])['value'].mean())
statesummary.head()
statesummary=statesummary.reset_index()

statesummary.head()
statesummary.pivot(index='State',columns='variable')
```

```python
merge_final_df = pd.merge(statesummary, Avg_asthma_noUS, on =['State'])
merge_final_df
```

```python
pvt_merged=pd.DataFrame(merge_final_df.pivot_table(index=['State','Asthma Mortality'], columns='variable'))
pvt_merged.columns = ['_'.join(tup).rstrip('_') for tup in pvt_merged.columns.values]
pvt_merged.reset_index(inplace=True)
pvt_merged.to_csv('asthma_polluts_avg_2010_16.csv', index=True, header=True)

pvt_merged.head()
#merge_final_df.columns = merge_final_df.columns.droplevel().rename(None)
#pvt_merged.reset_index(inplace=True)
# pvt_merged.reset_index(inplace=True)
# pvt_merged.reset_index(inplace=True)

#pvt_merged.reset_index(drop=True)

#merge_final_df.reset_index().fillna("null").to_csv("test.csv", sep="\t", index=None)
```
